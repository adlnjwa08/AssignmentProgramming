2. Print the first 20 prime numbers in a text file primes.txt. Then add another 30 prime numbers to primes.txt

```python
def is_prime(n):
    if n < 2: return False 
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

primes_list = []
num = 2
while len(primes_list) < 20:
    if is_prime(num):
        primes_list.append(num)
    num += 1

with open('primes.txt', 'w') as file:
    for p in primes_list:
        file.write(f'{p} ')
```

6.(c) Calculate the grade of each student (score ≥ 80 for A, 60 ≤ score < 80 for B, 40 ≤ score < 60
for C, and score < 40 for E). Write all data in students.csv to student_grade.csv and add the
grades as the fourth column of student_grade.csv.

```python
import csv 
def findGrade(score):
    if score >= 80:
        grade = 'A'
    elif score >= 60:
        grade = 'B'
    elif score >= 40:
        grade = 'C'
    else:
        grade = 'E'
    return grade
with open('student_grade.csv', 'w', newline='') as file_c:
    writer_c = csv.writer(file_c)
    writer_c.writerow(['Name', 'ID', 'Scores', 'Grade'])

    for row in all_data:
        name = row[0]
        id_num = row[1]
        score = int(row[2])
        grade = findGrade(score)
        writer_c.writerow([name, id_num, score, grade])
```

11.(a) searchByID that takes a staff ID as input and searches the staff.csv file for a staff member with
that ID. The function should display the details of the staff member if found. If no match is
found, it should show a message indicating that no staff member with that ID was found.

```python
def searchByID(staff_id):
    with open('staff.csv', 'r') as file:
        reader = csv.reader(file)
        next(reader)
        for row in reader:
            if row[1] == staff_id:
                print(f'Found: {row}')
                return
        print(f'No staff member with ID {staff_id} was found.')
searchByID('ST52')
searchByID('ST60')
searchByID('ST00')
```

11.(b) searchbyDepartment that takes a department name as input and searches the staff.csv file for
staff members who work in that department. The function should display the details of all staff
members found in that department. If no staff members are found, it should show a message
indicating that no staff members work in that department.

```python
def searchbyDepartment(dept_name):
    found_any = False
    with open('staff.csv','r') as file:
        reader = csv.reader(file)
        next(reader)
        for row in reader:
            if row[2].lower() == dept_name.lower():
                print(row)
                found_any = True
    if found_any == False:
        print(f'No staff members work in the {dept_name} department.')

searchbyDepartment('IT')
searchbyDepartment('Sales')
searchbyDepartment('Marketing')
```

12.(c) deleteContact that deletes a selected contact from contacts.csv. Let the user select the index
of the contact to be deleted, for example, if the contacts are displayed as
1. John 011 -2687357
2. Tom 012 -4573695
3. William 015 -2455873

If the user selects 2, then Tom will be removed from contacts.csv.

```python
def deleteContact(index):
    with open('contacts.csv','r') as file:
        rows = list(csv.reader(file))
    if 0 <= index + 1 < len(rows) - 1:
        deleted = rows.pop(index + 1)
        with open('contacts.csv','w', newline='') as file:
            csvwriter = csv.writer(file)
            csvwriter.writerows(rows)
        print(f'Deleted contact: {deleted[0]}')
    else:
        print('Invalid index.')

deleteContact(4)
deleteContact(3)
```

14. Creates a csv file fruits.csv that contain the following data. 

Fruit Price (RM) Expiry date
Kiwi 2.50 5/4/2024
Apple 5.50 30/5/2024
Orange 5.25 30/4/2024
Pear 4.70 15/5/2024
Durian 10.00 15/7/2024

Use csvDictReader to read the data into a dictionary. In the dictionary, add another field that specifies
the condition of the fruit: 'good' if the fruit is not yet passed its expiry date or 'expired' of the fruit
already passed its expiry date. Then print your output in a csv file fruitExpired.csv

```python
import csv
with open('fruits.csv','w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(['Fruit','Price (RM)','Expiry Date'])
    csvwriter.writerow(['Kiwi','2.50','5/4/2024'])
    csvwriter.writerow(['Apple','5.50','30/5/2024'])
    csvwriter.writerow(['Orange','5.25','30/4/2024'])
    csvwriter.writerow(['Pear','4.70','15/5/2024'])
    csvwriter.writerow(['Durian','10.00','15/7/2024'])
```

```python
import csv
from datetime import datetime

today = datetime.now()
processed_data = []

with open('fruits.csv','r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        expiry = datetime.strptime(row['Expiry Date'], '%d/%m/%Y')
        
        if expiry < today:
            row['Condition'] = 'expired'
        else:
            row['Condition'] = 'good'
            
        processed_data.append(row)

headers = ['Fruit', 'Price (RM)', 'Expiry Date', 'Condition']
with open('fruitExpired.csv', 'w', newline='') as f:
    csvwriter = csv.DictWriter(f,fieldnames=headers)
    csvwriter.writeheader()
    csvwriter.writerows(processed_data)

for fruit in processed_data:
    print(fruit)
```
