### QUESTION 5

```python
def is_leap_year(year):
    if year % 4 == 0:
        return False
    else:
        if year % 400 == 0:
            return True
        elif year % 100 == 0:
            return False
        else:
            return True
year = int(input("Enter a year: "))
if is_leap_year(year):
    print (f"{year} is a leap year.")
else:
    print (f"{year} is not a leap year.")
```

### QUESTION 12

```python
students = {
    "Frank": 65,
    "Annie": 52,
    "Ben": 71,
    "Tommy": 80,
    "Jenson": 99,
    "Lisa": 25
}
def get_grade(score):
    if score>=80:
        return ('A')
    elif score>=60:
        return ('B')
    elif score>=50:
        return ('C')
    elif score>=40:
        return ('D')
    else:
        return ('F')

print("Student Grades:")
print("-" * 30)

for name, score in students.items():
    grade = get_grade(score)
    print(f"Name: {name}, Score: {score}, Grade: {grade}")
```

### QUESTION 14

```python
n = int(input("Enter the number of Fibonacci numbers to generate: "))
fib_list = []
if n>=1:
    fib_list.append(0)
if n>=2:
    fib_list.append(1)
for i in range(2,n):
    next_fib = fib_list[i-1] + fib_list[i-2]
    fib_list.append(next_fib)
print (f"The first {n} Fibonacci numbers are: {fib_list}")
```

```python
#(c)
n = 7
print("Pattern 1")

for a1 in range(1, (n+1)//2 + 1):
    for a2 in range((n+1)//2 - a1):
        print(" ", end = "")
    for a3 in range(a1):
        print("* ", end = "")
    print()

for a1 in range((n+1)//2, 0, -1):
    for a2 in range((n+1)//2 - a1 + 1):
        print(" ", end = "")
    for a3 in range(a1 - 1):
        print ("* ", end = "")
    print()
```

### QUESTION 18

```python
import random 
def play_game():
    number = random.randint(1,1000)
    guess_count = 0
    guessed_correctly = False

    print("I'm thinking of a number between 1 and 1000. Can you guess it?")

    while not guessed_correctly:
        try:
            guess = int(input("Enter your guess: "))
            guess_count += 1

            if guess > number:
                print("Too high!")
            elif guess < number:
                print ("Too low!")
            else:
                guessed_correctly = True
            print(f"Congratulations! You passed it in {guess_count} guesses.")

            if guess_count == 1:
                print("Magnificent! Either you are lucky or you are really good at this.")
            elif guess_count <= 10:
                print("You really know how to play this game.")
            else:
                print("Try strategize your guess next time.")
        except ValueError:
            print("Please enter a valid integer.")

def main():
    play_again = 'y'
    while play_again == 'y':
        play_game()
        play_again = input("Do you want to play again? (y/n): ").lower().strip()
        if play_again not in ['y', 'n']:
            print("Invalid input, Assuming no. ")
            play_again = 'n'

if __name__ == "__main__":
    main()
```
