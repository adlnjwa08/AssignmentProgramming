(a) Write a function rangeOfNumbers that can accept numeric inputs of any length and returns the
range of the inputs 
(the difference between the largest input and the smallest input)

```python
def rangeOfNumbers(*args):
    if not args:
        return 0
    return max(args) - min(args)

print(rangeOfNumbers(10, 2, 5, 20))
```

(b) Write a function partialSum that can accept numeric inputs of any length and returns the list of
the partial sum of the inputs.

```python
def partialSum(*args):
    result = []
    current_sum = 0
    for num in args:
        current_sum += num
        result.append(current_sum)
    return result

print(partialSum(1, 2, 3, 4))
```

(c) Write a function that accepts keyword-value inputs of any length and prints the key-value pair in
tabular format.

```python
def print_tabular(**kwargs):
    print(f"{'Key':<10} | {'Value':<10}")
    print("-" * 23)
    for key, value in kwargs.items():
        print(f"{key:<10} | {value:<10}")

print_tabular(Name="Maria", Age=22, City="Italy")
```

(d) Write a function that accepts keyword-value inputs of any length. The keys are strings that
represents item names (like eggs, bread, milk, etc) and the value is a 2-element tuple of the
quantity and the price for that item. Calculate the total price for each item and the total of all
items and return the result in a dictionary. A sample function call looks like this
f(eggs=(1,2.5),bread=(3,3.6),milk=(4,2.25),rice=(2,5.7))
and the output looks like this

{'eggs': 2.5, 'bread': 10.8, 'milk': 9.0, 'rice': 11.4, 'total': 33.7}

```python
def calculate_grocery_total(**items):
    report = {}
    grand_total = 0
    
    for item, (quantity, price) in items.items():
        subtotal = quantity * price
        report[item] = subtotal
        grand_total = subtotal
    
    report['total'] = grand_total
    return report

result = calculate_grocery_total(eggs=(1, 2.5), bread=(3, 3.6), milk=(4, 2.25), rice=(2, 5.7))
print(result)
```

Write a Python function that

 (a) Write a recursive function that returns the sum of the first n positive integers.

```python
def sum_n(n):
    if n <= 0:
        return 0
    else:
        return n + sum_n(n - 1)
print("Sum of first 5 positive integers:", sum_n(5))
```

(b) Write a recursive function that returns the sum of all digits in an integer.

```python
def sum_digits(n): 
    n = abs(n) 
    if n < 10:
       return n
    else: 
        return n % 10 + sum_digits(n//10)
print("Sum of digits in 1234:", sum_digits(1234))
```

(c) Write a recursive function that returns the n-th Fibonacci number. The Fibonacci’s numbers are
generated by the following sequence

F(1) = 1
F(2) = 1
F(3) = F(2)+F(1)
F(4) = F(3)+F(2)
· · ·
F(n) = F(n−1)+F(n−2)

```python
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1 or n == 2:
         return 1
    else: 
        return fibonacci(n-1) + fibonacci(n-2) 

print ("7th Fibonacci number:", fibonacci(7))
```

(d) A Pascal’s triangle is a triangular array where the elements on the n-th row is the sum of the two
elements above it. A five row Pascal triangle looks like

6
1
1 1
1 2 1
1 3 3 1
1 4 6 4 1

Write a recursive function that returns the Pascal triangle of n rows,where n is the input(return
the result as a list of lists).All numbers can be aligned to the left.

```python
def pascal_triangle(n):
    if n<= 0:
        return []
    elif n == 1:
        return [[1]]
    else:
        triangle = pascal_triangle(n - 1)
        prev_row = triangle[-1]
        new_row = [1] 
        for i in range(len(prev_row) - 1):
            new_row.append(prev_row[i] + prev_row[i + 1])
        new_row.append(1)
        triangle.append(new_row)
        return triangle

rows=5
triangle = pascal_triangle(rows)
print(f"Pascal's Triangle with {rows} rows:")
for row in triangle:
    print(row)


```

```python
def fahrenheit_to_celsius(f):
    return (f - 32) * 5 / 9

fahrenheit_temps = [125.7, 275.4, 300.0, 37.48, 0.0, -180.45]


print("Using a loop:")
print(f"{'Fahrenheit':>12} | {'Celsius':>10}")
print("-" * 25)
for f in fahrenheit_temps:
    c = fahrenheit_to_celsius(f)
    print(f"{f:12.2f} | {c:10.2f}")


```
