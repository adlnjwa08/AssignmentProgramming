QUESTION 16 Solve the following system of equations

```python
import sympy as sym
x, y, z, t, r, s = sym.symbols('x y z t r s')
```

a) 10x + 12y + 16t = 0, 5x − y = 13t for

i) x and y in terms of t.

```python
eq1 = sym.Eq(10*x + 12*y + 16*t, 0)
eq2 = sym.Eq(5*x - y, 13*t)
sol_a1 = sym.solve([eq1, eq2], (x,y))
sol_a1
```

ii) t and y in terms of x.

```python
sol_a2 = sym.solve([eq1, eq2], (t, y))
sol_a2
```

iii) t and x in terms of y.

```python
sol_a3 = sym.solve([eq1, eq2], (t, x))
sol_a3
```

b) rx + sy + tz = 1, 2rx + sy − tz = 2,−rx + 2sy − 2tz = 3 for

i) x, y and z in terms of r, s and t.

```python
eq1 = sym.Eq(r*x + s*y + t*z, 1)
eq2 = sym.Eq(2*r*x + s*y - t*z, 2)
eq3 = sym.Eq(-r*x + 2*s*y - 2*t*z, 3)
sol_b1 = sym.solve([eq1, eq2, eq3], (x, y, z))
sol_b1
```

ii) r, s and t in terms of x, y and z.

```python
sol_b2 = sym.solve([eq1, eq2, eq3], (r, s, t))
sol_b2
```
